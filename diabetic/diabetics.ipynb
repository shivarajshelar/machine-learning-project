 ### KNN model

```python
x = df.drop(['Outcome'],axis=1)
y = df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.22,random_state=19)
```

```python
accuracy_train = []
accuracy_test = []
for i in range(1,21):
    clf = KNeighborsClassifier(n_neighbors=i)
    clf.fit(x_train,y_train)
    accuracy_train.append(clf.score(x_train,y_train))
    accuracy_test.append(clf.score(x_test,y_test))
    
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(x = range(1,21,1),y = accuracy_train )
sns.lineplot(x = range(1,21,1),y= accuracy_test)
plt.xlabel("neighbours numbers")
plt.ylabel("accuracy")
```

```python
clf = KNeighborsClassifier(n_neighbors=8)
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

### Accuracy 

```python
accuracy_score(y_test,y_pred)
```
